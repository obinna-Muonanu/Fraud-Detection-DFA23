```python
# import necessary libraries

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,StratifiedKFold
import lightgbm as lgb
from sklearn.metrics import accuracy_score, precision_score, f1_score,confusion_matrix,classification_report
```

```python
Train = pd.read_csv('/content/drive/MyDrive/Fraud Detection for Online payment Platform- Datafest Datathon/Feature Scaling.csv')
Train
```

```python
y = Train['Fraudulent Flag'] # Assigning target feature to y
```

```python
Train.drop(['Fraudulent Flag'],axis=1) # dropping target feature
```

```python
X_train, X_test, y_train, y_test = train_test_split(Train,y,test_size=0.1,random_state=5) # splitting data for training and testing
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
stratified_kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=5)
```

```python
p_scores = [] # list to hold precision scores
```

```python
# stratified kfold cross validation
# further split training data into training and validation sets
for train_index, test_index in stratified_kfold.split(X_train,y_train):
    train_x,test_x = X_train.iloc[train_index],X_train.iloc[test_index]
    train_y,test_y = y_train.iloc[train_index],y_train.iloc[test_index]

    model = lgb.LGBMClassifier()
    model.fit(train_x,train_y)

    target_pred = model.predict(test_x)
    score = precision_score(test_y,target_pred)
    p_scores.append(score)

p_scores
mean_score = np.mean(p_scores)
mean_score
```

```python
y_predictions = model.predict(X_test) # Predict X_test (unseen data)
```

```python
precision = precision_score(y_test,y_predictions)
precision
```

```python
confn_matrix = confusion_matrix(y_test,y_predictions)
confn_matrix
```

```python
accuracy_scores = [] # list to hold the accuracy scores
```

```python
for train_index, test_index in stratified_kfold.split(X_train,y_train):
    train_x,test_x = X_train.iloc[train_index],X_train.iloc[test_index]
    train_y,test_y = y_train.iloc[train_index],y_train.iloc[test_index]

    model = lgb.LGBMClassifier()
    model.fit(train_x,train_y)

    target_pred = model.predict(test_x)
    score = accuracy_score(test_y,target_pred)
    accuracy_scores.append(score)

accuracy_scores
mean_score = np.mean(accuracy_scores)
mean_score
```
