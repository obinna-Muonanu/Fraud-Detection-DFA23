**Import the necessary libraries**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler,StandardScaler,LabelEncoder
import copy
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
data = pd.read_csv('/content/drive/MyDrive/Fraud Detection for Online payment Platform- Datafest Datathon/Fraud Detection Dataset.csv')
data
```

```python
Data = copy.deepcopy(data)
Data.head()
```

```python
Data['Transaction Date and Time'] = pd.to_datetime(Data['Transaction Date and Time'])
Data.head()
```

```python
Data.info()
```

```python
Data.columns
```

**Feature Engineering**

```python
#Transactions per user per day
# To know how many transactions each user make per day
Data['Transactions/user/day'] = Data.groupby(['User ID',Data['Transaction Date and Time'].dt.date])['Transaction ID'].transform('count')
Data.head()
```

```python
Data['Day_of_Week'] = pd.to_datetime(Data['Transaction Date and Time']).dt.dayofweek
Data['Hour_of_Day'] = pd.to_datetime(Data['Transaction Date and Time']).dt.hour
Data.head()
```

```python
Data['Transaction_amount_to_income_ratio'] = Data['Transaction Amount']/Data['User Income']
Data.head()
```

```python
#Number of different authentication methods used by a user
Data['Number_of_Authentication_method_used'] = Data.groupby(['User ID'])['Transaction Authentication Method'].transform('nunique')
Data.tail()
```

```python
# Interaction between a user's age and transaction amount

Data['Age_x_Amount'] = Data['User Age'] * Data['Transaction Amount']
Data.head()
```

```python
# Number of unique IP address associated with a user/merchant
Data['Num_of_unique_IP_Address'] = Data.groupby(['User ID','Merchant ID'])['IP Address'].transform('nunique')
Data.head()
```

```python
#Frequency of transaction types for each user/merchant
Data['User_Transaction_Type_Frequency'] = Data.groupby(['User ID', 'Transaction Type'])['Transaction ID'].transform('count')
Data['Merchant_Transaction_Type_Frequency'] = Data.groupby(['Merchant ID', 'Transaction Type'])['Transaction ID'].transform('count')
Data
```

**Encoding Categorical Features**

```python
label_encoder = LabelEncoder()
```

```python
Data['Payment Method encoded'] = label_encoder.fit_transform(Data['Payment Method'])
Data['Country encoded'] = label_encoder.fit_transform(Data['Country Code'])
Data['Transaction Type encoded'] = label_encoder.fit_transform(Data['Transaction Type'])
Data['Device Type encoded'] = label_encoder.fit_transform(Data['Device Type'])
Data['Browser Type encoded'] = label_encoder.fit_transform(Data['Browser Type'])
Data['Operating System encoded'] = label_encoder.fit_transform(Data['Operating System'])
Data['Merchant Category encoded'] = label_encoder.fit_transform(Data['Merchant Category'])
Data['User Gender encoded'] = label_encoder.fit_transform(Data['User Gender'])
Data['User Occupation encoded'] = label_encoder.fit_transform(Data['User Occupation'])
Data['User Account Status encoded'] = label_encoder.fit_transform(Data['User Account Status'])
Data['Transaction Status encoded'] = label_encoder.fit_transform(Data['Transaction Status'])
Data["User's Device Location encoded"] = label_encoder.fit_transform(Data["User's Device Location"])
Data['Transaction Currency encoded'] = label_encoder.fit_transform(Data['Transaction Currency'])
Data['Transaction Purpose encoded'] = label_encoder.fit_transform(Data['Transaction Purpose'])
Data["User's Email Domain encoded"] = label_encoder.fit_transform(Data["User's Email Domain"])
Data["Transaction Authentication Method"] = label_encoder.fit_transform(Data["Transaction Authentication Method"])
Data['Transaction Time of Day encoded'] = label_encoder.fit_transform(Data['Transaction Time of Day'])
```
