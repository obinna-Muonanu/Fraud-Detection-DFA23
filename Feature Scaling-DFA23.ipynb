```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
my_Data = pd.read_csv("/content/drive/MyDrive/Fraud Detection for Online payment Platform- Datafest Datathon/Feature Engineering.csv")
my_Data
```

```python
# Feature scaling(normalization) numerical features

scaler = MinMaxScaler()
my_Data["User ID"] = scaler.fit_transform(my_Data[["User ID"]])
my_Data["Transaction Amount"] = scaler.fit_transform(my_Data[["Transaction Amount"]])
my_Data["Merchant ID"] = scaler.fit_transform(my_Data[["Merchant ID"]])
my_Data["User Age"] = scaler.fit_transform(my_Data[["User Age"]])
my_Data["User Income"] = scaler.fit_transform(my_Data[["User Income"]])
my_Data["Location Distance"] = scaler.fit_transform(my_Data[["Location Distance"]])
my_Data["Time Taken for Transaction"] = scaler.fit_transform(my_Data[["Time Taken for Transaction"]])
my_Data["User's Transaction History"] = scaler.fit_transform(my_Data[["User's Transaction History"]])
my_Data["Merchant's Reputation Score"] = scaler.fit_transform(my_Data[["Merchant's Reputation Score"]])
my_Data["User's Credit Score"] = scaler.fit_transform(my_Data[["User's Credit Score"]])
my_Data["Merchant's Business Age"] = scaler.fit_transform(my_Data[["Merchant's Business Age"]])
```
